```python



# Example usage
csv_file_path = '/home/ubuntu/Dawood/Augmentation/activity_dataset/merged_train_annotations.csv'  # Replace with your input CSV file path
output_file_path = '/home/ubuntu/Dawood/Augmentation/activity_dataset/normalized_merged_train_annotations.csv'  # Replace with your desired output file path
```

```python
import pandas as pd
import cv2
import os

# Define the column names
columns = ['videoname', 'frame', 'x1', 'y1', 'x2', 'y2', 'class', 'track_id']


def get_image_dimensions(video_name):
    base_path = '/home/ubuntu/Dawood/activity_dataset/20231004-activity-all-annotated-tasks'
    folder_path = os.path.join(base_path,video_name)
    # Assumes the first image is representative of all frame sizes
    first_image_path = os.path.join(folder_path, sorted(os.listdir(folder_path))[1])
    image = cv2.imread(first_image_path)
    return image.shape[:2]

# Read the CSV file without headers
df = pd.read_csv(csv_file_path, header=None, names=columns)

# Process each row
for index, row in df.iterrows():
    video_name = row['videoname']
    image_width, image_height = get_image_dimensions(video_name)

    # Check if the coordinates are already normalized
    if row['x1'] > 1 or row['y1'] > 1 or row['x2'] > 1 or row['y2'] > 1:
        # Normalize the bounding box coordinates
        df.at[index, 'x1'] = round(row['x1'] / image_width, 3)
        df.at[index, 'y1'] = round(row['y1'] / image_height, 3)
        df.at[index, 'x2'] = round(row['x2'] / image_width, 3)
        df.at[index, 'y2'] = round(row['y2'] / image_height, 3)

# Save the normalized annotations back to a new CSV file
df.to_csv(output_file_path, index=False)
```
